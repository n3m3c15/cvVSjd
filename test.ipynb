```python
import PyPDF2
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import SentenceTransformersTokenTextSplitter
from prompt import RESUME_PATH,JD_PATH


print(RESUME_PATH)
print(JD_PATH)

loader = PyPDFLoader(file_path=RESUME_PATH)
documents = loader.load()
text_splitter = CharacterTextSplitter()
documents_split = text_splitter.split_documents(documents=documents)
# print(documents)
texts = []
for doc in documents:
    page_content = doc.page_content
    texts.append(page_content)
# print(texts)
    
full_resume = ""
for text in texts:
    full_resume+=text
```

```python
from langchain_ollama import OllamaLLM
from colorama import Fore
import json

llm = OllamaLLM(model='deepseek-r1:7b')

messages = [{
    'role': 'system',
    'content':f'You are given a copy of a job descripiton = {JD_PATH} and my resume = {full_resume}...based on the provided documents give me completely factual, unbiased answers to my queries. You can be stern and ruthless with your answers tough love is appreciated'
}]

while True:
    query = input('User: ')
    if query != '/quit':
        messages.append(dict(role = 'user', content = query))
        response = llm.invoke(f"query = {query}, chat_history = {messages}")
        for chunk in response:
            print(Fore.GREEN + chunk, end="", flush=True)
        print(Fore.RESET)
        messages.append(dict(role = 'Bot', content = response.content))
    else:
        with open(file='messages_0.json',mode="+w") as f:
            json.dump(messages, f, indent=4)
        print("It was nice talking to you ヾ(￣▽￣) Bye~Bye~")
        break
```
